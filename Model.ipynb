```python
# importing libraries

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# LHS

```python
# Company X Data

xdf1 = pd.read_excel('Company X - Order Report.xlsx')
xdf2 = pd.read_excel('Company X - Pincode Zones.xlsx')
xdf3 = pd.read_excel('Company X - SKU Master.xlsx')
display(xdf1,xdf2,xdf3)
```

```python
# merging dataframes based on [SKU] column

xdf = pd.merge(xdf1,xdf3,on=['SKU'],how='left')
xdf
```

```python
# here while merging due to unknow reason 1 duplicate value is being created so to check that

print(xdf1.ExternOrderNo.value_counts())
print(xdf.ExternOrderNo.value_counts())
```

```python
# to find that duplicate values

display(xdf1.loc[xdf1.ExternOrderNo == 2001827036])
display(xdf.loc[xdf.ExternOrderNo == 2001827036])
```

```python
#now to get indices of this duplicate values

xdf.loc[(xdf.ExternOrderNo == 2001827036) & (xdf.SKU == 'GIFTBOX202002')]
```

```python
# droping that duplicate value

xdf = xdf.drop(5,axis=0)
xdf.loc[xdf.ExternOrderNo == 2001827036]
```

```python
# to calculate [Total Weight as per X (KG)]

xdf['Total Weight as per X (KG)'] = (xdf['Order Qty']*xdf['Weight (g)'])/1000
xdf
```

```python
# checking maximum weight

xdf['Total Weight as per X (KG)'].max()
```

```python
# slab weight calculations as per company X

def convert_to_slab(x):
    if x < 0.5:
        x = 0.5
    elif (x > 0.5) & (x < 1):
        x = 1
    else:
        x = 1.5
    return x

xdf['Weight Slab as per X (KG)'] = xdf['Total Weight as per X (KG)']
xdf['Weight Slab as per X (KG)'] = xdf['Weight Slab as per X (KG)'].apply(convert_to_slab)
xdf
```

```python
# calculating total number or [Order QTY], [Weight (g)] and [Total Weight as per X (KG)] for each unique ExternOrderNo

xdf = xdf.groupby("ExternOrderNo").sum()
xdf = xdf.reset_index()
xdf
```

```python
# droping unwanted columns

xdf.rename({'ExternOrderNo':'Order ID'},axis=1,inplace=True)
xdf.drop(['Order Qty','Weight (g)'],axis=1,inplace=True)
xdf
```

```python
# as company X data is not suffient to determine [Zone] and [Type of Shipment]
# so that data will be collected from Courier Company Invoice

cdf1 = pd.read_excel('Courier Company - Invoice.xlsx')
cdf1 = cdf1.drop(['Charged Weight','Warehouse Pincode','Billing Amount (Rs.)','Customer Pincode','AWB Code'],axis=1)
cdf1
```

```python
# merging dataframes

xdf = pd.merge(xdf,cdf1)
xdf
```

```python
# checking maximum slab weight

xdf['Weight Slab as per X (KG)'].max()
```

```python
# importing Courier Company Rates

cdf2 = pd.read_excel('Courier Company - Rates.xlsx')
cdf2
```

```python
# calculating Expected Charges courier company should has charged

xdf['Expected Charge as per X (Rs.)'] = 1000

list1 = np.arange(0.5,7.5,0.5)
list2 = np.arange(0,14)

for i in range(0,(len(xdf.Zone))):
    for j in range(0,len(list1)):
        if xdf['Weight Slab as per X (KG)'][i] == list1[j]:
            if xdf.Zone[i] == 'a':
                xdf['Expected Charge as per X (Rs.)'][i] = cdf2.fwd_a_fixed+(cdf2.fwd_a_additional*list2[j])
            elif xdf.Zone[i] == 'b':
                xdf['Expected Charge as per X (Rs.)'][i] = cdf2.fwd_b_fixed+(cdf2.fwd_b_additional*list2[j])
            elif xdf.Zone[i] == 'c':
                xdf['Expected Charge as per X (Rs.)'][i] = cdf2.fwd_c_fixed+(cdf2.fwd_c_additional*list2[j])
            elif xdf.Zone[i] == 'd':
                xdf['Expected Charge as per X (Rs.)'][i] = cdf2.fwd_d_fixed+(cdf2.fwd_d_additional*list2[j])
            else:
                xdf['Expected Charge as per X (Rs.)'][i] = cdf2.fwd_e_fixed+(cdf2.fwd_e_additional*list2[j])
                
xdf
```

```python
# as we have take Zone data from Courier Company Invoice ther will be no changes in that Data

xdf.rename({'Zone':'Delivery Zone as per X'},axis=1,inplace=True)
xdf
```

```python
# finalising the LHS

xdf = xdf.drop(['Type of Shipment'],axis=1)
xdf
```

# RHS

```python
# importing Courier company Invocie

cdf1 = pd.read_excel('Courier Company - Invoice.xlsx')
cdf1
```

```python
# slab weight calculations as per Courier Company


def convert_to_slab_1(x):
    if x < 0.5:
        x = 0.5
    elif (x > 0.5) & (x < 1):
        x = 1
    elif (x > 1) & (x < 1.5):
        x = 1.5
    elif (x > 1.5) & (x < 2):
        x = 2
    elif (x > 2) & (x < 2.5):
        x = 2.5
    elif (x > 2.5) & (x < 3):
        x = 3
    elif (x > 3) & (x < 3.5):
        x = 3.5
    elif (x > 3.5) & (x < 4):
        x = 4
    else:
        x = 4.5
    return x

cdf1['Weight slab as per Courier Company (KG)'] = cdf1['Charged Weight']
cdf1['Weight slab as per Courier Company (KG)'] = cdf1['Weight slab as per Courier Company (KG)'].apply(convert_to_slab_1)
cdf1
```

```python
# droping, renaming and finalising RHS

cdf = cdf1.drop(['Warehouse Pincode','Type of Shipment','Customer Pincode'],axis=1)
cdf.rename({'Charged Weight':'Total weight as per Courier Company (KG)','Zone':'Delivery Zone as per Courier Company','Billing Amount (Rs.)':'Charges Billed by Courier Company (Rs.)','AWB Code':'AWB Number'},axis=1,inplace=True)
cdf
```

# Merging Both Sides

```python
# merging LHS and RHS

df1 = pd.merge(xdf,cdf,on='Order ID',how='inner')
df1
```

```python
# rearanging columns

df1 = df1[['Order ID','AWB Number','Total Weight as per X (KG)','Weight Slab as per X (KG)','Total weight as per Courier Company (KG)','Weight slab as per Courier Company (KG)','Delivery Zone as per X','Delivery Zone as per Courier Company','Expected Charge as per X (Rs.)','Charges Billed by Courier Company (Rs.)']]
df1
```

```python
# SHEET 1 : "Calculation"
# final calculation of Difference between (Expected Charges) and (Billed Charges)

df1['Difference Between Expected Charges and Billed Charges (Rs.)'] = df1['Charges Billed by Courier Company (Rs.)'] - df1['Expected Charge as per X (Rs.)']
df1
```

```python
# Creating Data for Summary Sheet

Total_Orders_Correctly_Charged = []
Total_Orders_Over_Charged = []
Total_Orders_Under_Charged = []

for i in df1['Difference Between Expected Charges and Billed Charges (Rs.)']:
    if i == 0:
        Total_Orders_Correctly_Charged.append(i)
    elif i > 0:
        Total_Orders_Over_Charged.append(i)
    else:
        Total_Orders_Under_Charged.append(i)
        
sdf1 = pd.DataFrame({'Total Orders Correctly Charged':Total_Orders_Correctly_Charged})
sdf2 = pd.DataFrame({'Total Orders Over Charged':Total_Orders_Over_Charged})
sdf3 = pd.DataFrame({'Total Orders Under Charged':Total_Orders_Under_Charged})
display(sdf1,sdf2,sdf3)
```

```python
# merging data

sdf1 = sdf1.T
sdf2 = sdf2.T
sdf1 = sdf1.append(sdf2).T
sdf1 = sdf1.T
sdf3 = sdf3.T
sdf1 = sdf1.append(sdf3).T
sdf1
```

```python
# SHEET 2 : "Summary"
# calculation of Correctly, Over and Under charges

df2 = pd.DataFrame({'Count':sdf1.count(),'Amount':sdf1.sum()})
df2
```
